```python
import numpy as np
```

```python
X = np.random.normal(loc=10, scale=10, size=(100, 10))
print("Матрица X:\n", X)
```

```python
mean = np.mean(X, axis=0)
std_dev = np.std(X, axis=0)
normX = (X - mean) / std_dev

print("\nНормированная матрица:\n", normX)
```

```python
Z = np.array([[4, 5, 0],
              [1, 9, 3],
              [5, 1, 1],
              [3, 3, 3],
              [9, 9, 9],
              [4, 7, 1]])

row_sums = np.sum(Z, axis=1)
rows_gt_10 = np.nonzero(row_sums > 10)

print("\nИндексы строк, где сумма > 10:\n", rows_gt_10)
```

```python
A = np.eye(3)
B = np.eye(3)
AB = np.vstack((A, B))

print("\nМатрица A:\n", A)
print("\nМатрица B:\n", B)
print("\nОбъединённая матрица AB:\n", AB)
```
