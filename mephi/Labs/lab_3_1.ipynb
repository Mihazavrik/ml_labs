Импорт библиотек

```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Загрузка данных

```python
iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=241)
```

Обучение модели

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

Предсказание и оценка точности

```python
y_pred = clf.predict(X_test)
print("Accuracy (Decision Tree):", accuracy_score(y_test, y_pred))  # Доля верных предсказаний
```

Важность признаков

```python
importances = clf.feature_importances_
print("Feature importances:", importances)
```

Визуализация исходных меток и предсказанных

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pca_2d = X[:, :2]  # Используем 2 признака для визуализации

plt.figure(figsize=(10, 4))

# Реальные метки
plt.subplot(1, 2, 1)
sns.scatterplot(x=pca_2d[:, 0], y=pca_2d[:, 1], hue=iris.target)
plt.title("Реальные метки")

# Предсказания дерева решений
plt.subplot(1, 2, 2)
sns.scatterplot(x=pca_2d[:, 0], y=pca_2d[:, 1], hue=clf.predict(X))
plt.title("Предсказания дерева решений")

plt.tight_layout()
plt.show()
```
